```python
needed_fund_code = df_merged[df_merged['基金分类'].isin(needed)]['基金代码'].to_list()
needed_fund_code
```

```python
def split_list_into_batches(lst, num_batches):
    """
    将一个列表分为指定数量的批次
    :param lst: 要分割的列表
    :param num_batches: 批次数量
    :return: 一个包含批次的列表
    """
    # 计算每个批次的大小
    avg_batch_size = len(lst) // num_batches
    # 计算有多少元素是不能均分的
    remainder = len(lst) % num_batches

    batches = []
    start = 0

    for i in range(num_batches):
        # 当前批次的结束位置
        end = start + avg_batch_size + (1 if i < remainder else 0)
        # 将当前批次加入结果
        batches.append(lst[start:end])
        # 更新开始位置
        start = end

    return batches


split_list_into_batches(needed_fund_code, 100)

```
